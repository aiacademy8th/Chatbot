05. 랭그래프로 구현하는 인지 이키텍처

- 5장에서는 LLM 애플리케이션 구축에 사용되는 다양한 아키텍처 패턴을 랭그래프로 구현하는 방법을 설명, 자율성과 신뢰성 사이의 트레이트오프를 중심으로, 세 가지 인지 아키텍처를 소개하고 이를 코드로 구현, 단순한 LLM 호출부터 정해진 순서로 여러 LLM 을 호출하는 체인, 그리고 LLM이 다음 단계를 결정하는 라우터까지 자율성이 점진적으로 높아지는 아키텍처를 이용해 목적에 맞는 애플리케이션 설계 방법을 제시
